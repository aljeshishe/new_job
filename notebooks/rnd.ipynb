```python
from pathlib import Path
import dtale
import pandas as pd
import rootpath

```

```python
ROOT_PATH = Path(rootpath.detect())
df = pd.read_json(ROOT_PATH / "glassdoor/results/items_2023-02-01T10-42-14.json", lines=True)
columns = {"EmployerReviews_employer_shortName": "name",
           "EmployerReviews_employer_sizeCategory": "size",
           "EmployerJobs_jobsCount": "jobs",
           "EmployerSalary_salaryCount": "salaries",
           "EmployerReviews_allReviewsCount": "reviews",
           "EmployerReviews_employer_headquarters": "hq",
           "EmployerReviews_ratings_overallRating": "overall_rating",
           "EmployerReviews_ratings_careerOpportunitiesRating": "career_opportunities_rating",
           "EmployerReviews_ratings_compensationAndBenefitsRating": "compensation_and_benefits_rating",
           "EmployerReviews_ratings_cultureAndValuesRating": "culture_and_values_rating",
           "EmployerReviews_ratings_diversityAndInclusionRating": "diversity_and_inclusion_rating",
           "EmployerReviews_ratings_seniorManagementRating": "senior_management_rating",
           "EmployerReviews_ratings_workLifeBalanceRating": "work_life_balance_rating",
           "EmployerLocations_officeAddresses_administrativeAreaName1_list": "areas",
           "EmployerLocations_officeAddresses_cityName_list": "cities",
           "EmployerReviews_employer_links_overviewUrl": "url",
           "EmployerReviews_employer_overview_description": "description",
           }
df1 = pd.DataFrame()
for old, new in columns.items():
    df1[new] = df[old]

df1["url"] = df1.apply(lambda x: "https://www.glassdoor.com" + x["url"], axis=1)
df1
```
